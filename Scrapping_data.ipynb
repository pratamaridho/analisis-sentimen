```python
import pandas as pd
from google_play_scraper import reviews, Sort

APP_ID = "com.tokopedia.tkpd"
JUMLAH_REVIEW = 10000


print(f"Memulai scraping {JUMLAH_REVIEW} review untuk {APP_ID}...")

result, continuation_token = reviews(
    APP_ID,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=JUMLAH_REVIEW,
    filter_score_with=None
)

print(f"Berhasil mengambil {len(result)} review.")
df = pd.DataFrame(result)

# --- Menampilkan Hasil ---
print("\nContoh 5 data pertama:")
print(df[['userName', 'at', 'score', 'content']].head())


NAMA_FILE = "reviews_tokopedia.csv"
df.to_csv(NAMA_FILE, index=False, encoding='utf-8-sig')
print(f"\nData lengkap telah disimpan ke {NAMA_FILE}")
```
